## 4. Epsilon-greedy action selection  📝 TODO

We must balance **exploration** (try new actions) and **exploitation** (use the best
known action). The epsilon-greedy rule:

- With probability **ε**, pick a **random** action.
- Otherwise (prob **1 − ε**), pick the action with the **highest Q-value** in this state:
  `argmax_a Q[s, a]`.


````python
# 📝 TODO — epsilon-greedy policy
def epsilon_greedy(Q, state, epsilon):
    """Return an action index using the epsilon-greedy rule."""
    if np.random.random() < epsilon:
        return np.random.randint(Q.shape[1])
    else:
        return int(np.argmax(Q[state]))


# quick sanity check: with epsilon=0 it must return the greedy action
Q_test = np.zeros((n_states, n_actions))
Q_test[0, 3] = 1.0
assert epsilon_greedy(Q_test, 0, epsilon=0.0) == 3
print("epsilon_greedy looks good ✅")
````

## 5. The Q-learning update — the heart of everything  📝 TODO

After taking action `a` in state `s`, observing reward `r` and next state `s'`, we
update our estimate toward the **TD target**:

$$
Q(s,a) \;\leftarrow\; Q(s,a) \;+\; \alpha \Big[\, \underbrace{r + \gamma \max_{a'} Q(s', a')}_{\text{TD target}} \;-\; Q(s,a) \,\Big]
$$

- **α (alpha)** — learning rate: how much we move toward the target.
- **γ (gamma)** — discount: how much we value future reward.
- The bracket is the **TD error**.

Because Q-learning bootstraps off `max_a' Q(s', a')` (the *best* next action, not the
one we actually take next), it is an **off-policy** algorithm. Keep that phrase in mind —
it comes back in Project 2 with the *target network*.

Implement a single update step below.

````python
# 📝 TODO — one Q-learning update
def q_learning_update(Q, s, a, r, s_next, done, alpha, gamma):
    """Apply the Bellman update to Q in place and return the new Q[s, a] value.

    If `done` is True there is no next state, so the TD target is just r.
    """
    if done:
        best_next = 0.0
    else:
        best_next = np.max(Q[s_next])

    td_target = r + gamma * best_next
    Q[s, a] = Q[s, a] + alpha * (td_target - Q[s, a])

    return Q[s, a]

print("Update function defined ✅")
````

## 6. The training loop  📝 TODO (small)

This loop is mostly provided. You only need to **wire in the three functions you just
wrote** at the marked spots, and fill the **epsilon decay** line.

We decay ε from `EPS_START` down to `EPS_MIN` exponentially, so the agent explores
early and exploits later.

````python
# Hyperparameters
# Safety: recreate env if notebook kernel was restarted
import gymnasium as gym
import numpy as np

env = gym.make("Taxi-v3", render_mode="ansi")
n_states = env.observation_space.n
n_actions = env.action_space.n

N_EPISODES = 20000
ALPHA      = 0.1
GAMMA      = 0.99
EPS_START  = 1.0
EPS_MIN    = 0.05
EPS_DECAY  = 0.9995   # multiplied each episode

Q = np.zeros((n_states, n_actions))
epsilon = EPS_START
rewards_per_episode = []

for ep in range(N_EPISODES):
    state, _ = env.reset()
    done = False
    total_reward = 0

    while not done:
        # (a) choose action ----------------------------------------------------
        action = epsilon_greedy(Q, state, epsilon)

        # (b) step the environment --------------------------------------------
        next_state, reward, terminated, truncated, _ = env.step(action)
        done = terminated or truncated

        # (c) learn ------------------------------------------------------------
        q_learning_update(
            Q=Q,
            s=state,
            a=action,
            r=reward,
            s_next=next_state,
            done=done,
            alpha=ALPHA,
            gamma=GAMMA
        )

        state = next_state
        total_reward += reward

    # (d) decay epsilon --------------------------------------------------------
    epsilon = max(EPS_MIN, epsilon * EPS_DECAY)

    rewards_per_episode.append(total_reward)
    if (ep + 1) % 2000 == 0:
        avg = np.mean(rewards_per_episode[-2000:])
        print(f"Episode {ep+1:5d} | avg reward (last 2000): {avg:6.2f} | epsilon: {epsilon:.3f}")

print("\nTraining done ✅")
````

## 7. Plot the learning curve  ✅ PROVIDED

A healthy run climbs from roughly **−200** (random flailing) up to about **+7 to +8**
(near-optimal — an optimal Taxi policy averages ~7.9).

````python
# ✅ PROVIDED — smoothed learning curve
def moving_average(x, w=200):
    return np.convolve(x, np.ones(w)/w, mode="valid")

plt.figure(figsize=(9, 4))
plt.plot(moving_average(rewards_per_episode), linewidth=2)
plt.axhline(7.9, color="green", ls="--", label="≈ optimal (7.9)")
plt.title("Taxi-v3 — Q-learning learning curve (200-episode moving average)")
plt.xlabel("Episode")
plt.ylabel("Total reward")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
````

## 8. Watch the trained agent  ✅ PROVIDED

Greedy rollout (ε = 0). You should see the taxi drive to the passenger, pick up,
drive to the destination, and drop off in a small number of steps.

````python
# ✅ PROVIDED — greedy evaluation
def evaluate(Q, n_eval=100, render_one=False):
    eval_env = gym.make("Taxi-v3", render_mode="ansi")
    returns, lengths = [], []
    for i in range(n_eval):
        s, _ = eval_env.reset()
        done, R, steps = False, 0, 0
        frames = []
        while not done:
            a = int(np.argmax(Q[s]))
            s, r, term, trunc, _ = eval_env.step(a)
            done = term or trunc
            R += r; steps += 1
            if render_one and i == 0:
                frames.append(eval_env.render())
        returns.append(R); lengths.append(steps)
    print(f"Average return over {n_eval} episodes: {np.mean(returns):.2f}")
    print(f"Average episode length          : {np.mean(lengths):.2f} steps")
    if render_one:
        import time
        from IPython.display import clear_output
        for f in frames:
            clear_output(wait=True); print(f); time.sleep(0.3)
    return np.mean(returns)

evaluate(Q, n_eval=100)
print("\n--- Watch one episode ---")
evaluate(Q, n_eval=1, render_one=True)
````

````python
# 📝 TODO — Section 9 Ablation Experiments

import numpy as np
import matplotlib.pyplot as plt
import gymnasium as gym

# Safety: recreate env if needed
env = gym.make("Taxi-v3", render_mode="ansi")
n_states = env.observation_space.n
n_actions = env.action_space.n


def epsilon_greedy(Q, state, epsilon):
    """Return an action index using the epsilon-greedy rule."""
    if np.random.random() < epsilon:
        return np.random.randint(Q.shape[1])
    return int(np.argmax(Q[state]))


def q_learning_update(Q, s, a, r, s_next, done, alpha, gamma):
    """Apply one Q-learning Bellman update."""
    best_next = 0.0 if done else np.max(Q[s_next])
    td_target = r + gamma * best_next
    Q[s, a] = Q[s, a] + alpha * (td_target - Q[s, a])
    return Q[s, a]


def train_q_learning(
    n_episodes=20000,
    alpha=0.1,
    gamma=0.99,
    eps_start=1.0,
    eps_min=0.05,
    eps_decay=0.9995
):
    Q = np.zeros((n_states, n_actions), dtype=float)
    epsilon = eps_start
    rewards = []

    for ep in range(n_episodes):
        state, _ = env.reset()
        done = False
        total_reward = 0

        while not done:
            action = epsilon_greedy(Q, state, epsilon)

            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated

            q_learning_update(
                Q=Q,
                s=state,
                a=action,
                r=reward,
                s_next=next_state,
                done=done,
                alpha=alpha,
                gamma=gamma
            )

            state = next_state
            total_reward += reward

        epsilon = max(eps_min, epsilon * eps_decay)
        rewards.append(total_reward)

    return Q, rewards


def moving_average(x, w=200):
    return np.convolve(x, np.ones(w) / w, mode="valid")


def evaluate_greedy(Q, n_eval=100):
    eval_env = gym.make("Taxi-v3", render_mode="ansi")
    returns = []

    for _ in range(n_eval):
        state, _ = eval_env.reset()
        done = False
        total_reward = 0

        while not done:
            action = int(np.argmax(Q[state]))
            state, reward, terminated, truncated, _ = eval_env.step(action)
            done = terminated or truncated
            total_reward += reward

        returns.append(total_reward)

    return np.mean(returns)


# Baseline
Q_base, rewards_base = train_q_learning(
    alpha=0.1,
    gamma=0.99,
    eps_start=1.0,
    eps_min=0.05,
    eps_decay=0.9995
)

# A. gamma = 0
Q_gamma0, rewards_gamma0 = train_q_learning(
    alpha=0.1,
    gamma=0.0,
    eps_start=1.0,
    eps_min=0.05,
    eps_decay=0.9995
)

# B. epsilon fixed at 1.0
Q_eps_fixed, rewards_eps_fixed = train_q_learning(
    alpha=0.1,
    gamma=0.99,
    eps_start=1.0,
    eps_min=1.0,
    eps_decay=1.0
)

# C. alpha = 1.0
Q_alpha1, rewards_alpha1 = train_q_learning(
    alpha=1.0,
    gamma=0.99,
    eps_start=1.0,
    eps_min=0.05,
    eps_decay=0.9995
)


# Print greedy evaluation results
print("Greedy evaluation after training:")
print(f"Baseline           : {evaluate_greedy(Q_base):.2f}")
print(f"A. gamma = 0       : {evaluate_greedy(Q_gamma0):.2f}")
print(f"B. epsilon fixed 1 : {evaluate_greedy(Q_eps_fixed):.2f}")
print(f"C. alpha = 1       : {evaluate_greedy(Q_alpha1):.2f}")


# Plot learning curves
plt.figure(figsize=(10, 5))
plt.plot(moving_average(rewards_base), label="Baseline: gamma=0.99, alpha=0.1, epsilon decay")
plt.plot(moving_average(rewards_gamma0), label="A: gamma=0")
plt.plot(moving_average(rewards_eps_fixed), label="B: epsilon fixed at 1.0")
plt.plot(moving_average(rewards_alpha1), label="C: alpha=1.0")
plt.axhline(7.9, color="green", linestyle="--", label="≈ optimal return")
plt.title("Taxi-v3 Q-learning ablation experiments")
plt.xlabel("Episode")
plt.ylabel("Total reward, 200-episode moving average")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
````

## 10. Stretch (optional) — SARSA vs Q-learning

Implement **SARSA**, the *on-policy* cousin. The only change is the TD target:

- Q-learning (off-policy):  `target = r + γ · max_a' Q(s', a')`

  actually chosen by epsilon-greedy in `s'`.

Train both and compare. (The famous payoff is on `CliffWalking-v0`, where SARSA

````python
# 📝 Optional Stretch — SARSA vs Q-learning

def sarsa_update(Q, s, a, r, s_next, a_next, done, alpha, gamma):
    """Apply one SARSA update."""
    next_value = 0.0 if done else Q[s_next, a_next]
    td_target = r + gamma * next_value
    Q[s, a] = Q[s, a] + alpha * (td_target - Q[s, a])
    return Q[s, a]


def train_sarsa(
    n_episodes=20000,
    alpha=0.1,
    gamma=0.99,
    eps_start=1.0,
    eps_min=0.05,
    eps_decay=0.9995
):
    Q = np.zeros((n_states, n_actions), dtype=float)
    epsilon = eps_start
    rewards = []

    for ep in range(n_episodes):
        state, _ = env.reset()
        done = False
        total_reward = 0

        action = epsilon_greedy(Q, state, epsilon)

        while not done:
            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated

            if not done:
                next_action = epsilon_greedy(Q, next_state, epsilon)
            else:
                next_action = 0

            sarsa_update(
                Q=Q,
                s=state,
                a=action,
                r=reward,
                s_next=next_state,
                a_next=next_action,
                done=done,
                alpha=alpha,
                gamma=gamma
            )

            state = next_state
            action = next_action
            total_reward += reward

        epsilon = max(eps_min, epsilon * eps_decay)
        rewards.append(total_reward)

    return Q, rewards


# Train Q-learning and SARSA
Q_qlearn, rewards_qlearn = train_q_learning()
Q_sarsa, rewards_sarsa = train_sarsa()

print("Greedy evaluation:")
print(f"Q-learning: {evaluate_greedy(Q_qlearn):.2f}")
print(f"SARSA     : {evaluate_greedy(Q_sarsa):.2f}")

plt.figure(figsize=(10, 5))
plt.plot(moving_average(rewards_qlearn), label="Q-learning")
plt.plot(moving_average(rewards_sarsa), label="SARSA")
plt.axhline(7.9, color="green", linestyle="--", label="≈ optimal return")
plt.title("Taxi-v3: Q-learning vs SARSA")
plt.xlabel("Episode")
plt.ylabel("Total reward, 200-episode moving average")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
````

**SARSA vs Q-learning.**
Both Q-learning and SARSA eventually learn a strong Taxi-v3 policy, and both curves approach the near-optimal reward line. Q-learning learns faster early in training because it is off-policy and updates toward the best possible next action using `max Q(s', a')`. SARSA learns more slowly because it is on-policy and updates toward the action actually chosen by epsilon-greedy, which may still include exploratory/random actions. In Taxi-v3, the final difference is small because the environment is deterministic and does not have a risky cliff-like structure where SARSA’s safer behavior would matter more.
